```python
import os
import arcpy
import pandas as pd
import numpy as np
import csv
from arcgis.features import GeoAccessor, GeoSeriesAccessor

arcpy.env.overwriteOutput = True
```

```python
# ** Ubicación y registro de archivos .gdb dentro de múltiples directorios

ruta_padre = r"D:\PUBLIC\PGAR\Resultados\4.PRO\PGAR\BackUP_Pomcas\POMCAS_Truncados"
nombre_directorios = os.listdir(ruta_padre)

rutas_workspace = []
for directorio in nombre_directorios:
    ruta_bd = os.path.join(ruta_padre,directorio)
    arcpy.env.workspace = ruta_bd
    for dataset in arcpy.ListDatasets():
        for capa in arcpy.ListFeatureClasses(feature_dataset=dataset):
            arcpy.management.TruncateTable(capa)
            print("Se trunca la capa {0} del dataset {1} de la BD {2}".format(capa, dataset, directorio))
```
