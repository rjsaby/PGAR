```python
# !TODO: Asignación Espacio de Trabajo DB POMCA
# ** Borrado de capas
arcpy.env.workspace = capa_pomca_analisis

print("Borrado Capas")
lista_capa_dataset = []
for feature in arcpy.ListFeatureClasses():
    lista_capa_dataset.append(feature)
    if len(lista_capa_dataset) == 0:
        print("Vacio")
    else:
        arcpy.Delete_management(feature)
        print(f" - Se borra capa {feature}")
```

```python
# !TODO: Asignación Espacio de Trabajo DB POMCA
arcpy.env.workspace = ruta_base_POMCA
lista_ruta_origen = []
lista_ruta_destino = []
print("Copia de capas")
print(" - Asignación de Lista")
for dataset in arcpy.ListDatasets():
    for feature in arcpy.ListFeatureClasses(feature_dataset=dataset):
        for key, value in diccionario_capas_POMCA_analisis.items():
            if value[1] == feature:
                ruta_capa_origen = os.path.join(arcpy.env.workspace,dataset,feature)
                lista_ruta_origen.append(ruta_capa_origen)
                nombre_capa_salida = value[1] + '__' + value[2].replace('_','')
                ruta_capa_salida = os.path.join(capa_pomca_analisis,nombre_capa_salida)
                lista_ruta_destino.append(ruta_capa_salida)
                
print(" - Copia de Capas")
for ruta_origen, ruta_destino in zip(lista_ruta_origen, lista_ruta_destino):
    arcpy.management.Copy(ruta_origen, ruta_destino)
    print(f" - Se copia la capa: {os.path.basename(ruta_origen)}")
```
