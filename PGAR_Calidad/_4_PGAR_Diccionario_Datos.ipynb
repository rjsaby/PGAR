Creación de Funciones

```python
def diccionario_con_datasets(directorio_base_datos):
    # ? Asignación de Variable
    arcpy.env.workspace = directorio_base_datos
    
    # ? Bluque con Dataset
    for dataset in arcpy.ListDatasets():
        for capa in arcpy.ListFeatureClasses(feature_dataset=dataset):
            for campo in arcpy.ListFields(dataset=capa):
                nombre_db = os.path.basename(arcpy.env.workspace)
                nombre_base_datos.append(nombre_db)
                nombre_dataset.append(dataset)
                nombre_capa.append(capa)
                nombre_campo.append(campo.name)
                alias.append(campo.aliasName)
                tipo_campo.append(campo.type)
                obligatorio.append(campo.isNullable)
                longitud.append(campo.length)
                precision.append(campo.precision)
                escala.append(campo.scale)
                es_requerido.append(campo.required)
                nombre_dominio.append(campo.domain)     
        print(f"DataSet: {dataset} - Capa: {capa} - Finalizado")
        
    lista_w_dataset = list(zip(nombre_base_datos,
                                    nombre_dataset, 
                                    nombre_capa, 
                                    nombre_campo,
                                    alias,
                                    tipo_campo,
                                    obligatorio,
                                    longitud,
                                    precision,
                                    escala,
                                    es_requerido,
                                    nombre_dominio))

    # ? Creación de DataFrames
    dataframe_diccionario_w_dataset = pd.DataFrame(lista_w_dataset, columns=['nombre_bd','nombre_dataset','nombre_capa','nombre_campo','alias','tipo_campo','obligatorio','longitud','precision','escala','es_requerido','nombre_dominio'])
    
    return dataframe_diccionario_w_dataset

def diccionario_sin_datasets(directorio_base_datos):
    # ? Asignación de Variable
    arcpy.env.workspace = directorio_base_datos

    # ? Bluque sin Dataset
    for capa in arcpy.ListFeatureClasses():
        for campo in arcpy.ListFields(dataset=capa):
            nombre_db = os.path.basename(arcpy.env.workspace)
            nombre_base_datos.append(nombre_db)
            sin_dataset = 'Sin DataSet'
            nombre_dataset.append(sin_dataset)
            nombre_capa.append(capa)
            nombre_campo.append(campo.name)
            alias.append(campo.aliasName)
            tipo_campo.append(campo.type)
            obligatorio.append(campo.isNullable)
            longitud.append(campo.length)
            precision.append(campo.precision)
            escala.append(campo.scale)
            es_requerido.append(campo.required)
            nombre_dominio.append(campo.domain)     
        print(f"DataSet: {sin_dataset} - Capa: {capa} - Finalizado")
        
    lista_wout_dataset = list(zip(nombre_base_datos,
                                    nombre_dataset, 
                                    nombre_capa, 
                                    nombre_campo,
                                    alias,
                                    tipo_campo,
                                    obligatorio,
                                    longitud,
                                    precision,
                                    escala,
                                    es_requerido,
                                    nombre_dominio))

    # ? Creación de DataFrames
    dataframe_diccionario_wout_dataset = pd.DataFrame(lista_wout_dataset, columns=['nombre_bd','nombre_dataset','nombre_capa','nombre_campo','alias','tipo_campo','obligatorio','longitud','precision','escala','es_requerido','nombre_dominio'])
    
    return dataframe_diccionario_wout_dataset

def diccionario_tablas(directorio_base_datos):

    arcpy.env.workspace = directorio_base_datos

    # ? Bluque sin Dataset
    for capa in arcpy.ListTables():
        for campo in arcpy.ListFields(dataset=capa):
            nombre_db = os.path.basename(arcpy.env.workspace)
            nombre_base_datos.append(nombre_db)
            sin_dataset = 'Sin DataSet'
            nombre_dataset.append(sin_dataset)
            nombre_capa.append(capa)
            nombre_campo.append(campo.name)
            alias.append(campo.aliasName)
            tipo_campo.append(campo.type)
            obligatorio.append(campo.isNullable)
            longitud.append(campo.length)
            precision.append(campo.precision)
            escala.append(campo.scale)
            es_requerido.append(campo.required)
            nombre_dominio.append(campo.domain)     
        print(f"DataSet: {sin_dataset} - Capa: {capa} - Finalizado")
        
    lista_tablas_dataset = list(zip(nombre_base_datos,
                                    nombre_dataset, 
                                    nombre_capa, 
                                    nombre_campo,
                                    alias,
                                    tipo_campo,
                                    obligatorio,
                                    longitud,
                                    precision,
                                    escala,
                                    es_requerido,
                                    nombre_dominio))

    # ? Creación de DataFrames
    dataframe_diccionario_tablas_dataset = pd.DataFrame(lista_tablas_dataset, columns=['nombre_bd','nombre_dataset','nombre_capa','nombre_campo','alias','tipo_campo','obligatorio','longitud','precision','escala','es_requerido','nombre_dominio'])
   
    return dataframe_diccionario_tablas_dataset
```
