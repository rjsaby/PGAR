```python
import os
import arcpy
import pandas as pd
import numpy as np
import csv
from arcgis.features import GeoAccessor, GeoSeriesAccessor

arcpy.env.overwriteOutput = True
```

```python
ruta_origen = r"D:\PUBLIC\PGAR\Resultados\4.PRO\PGAR\BackUP_Pomcas\POMCAS_Truncados"
ruta_destino = r"D:\PUBLIC\PGAR\Resultados\4.PRO\PGAR\gdb_pomcas\POMCA_Unificada.gdb"
nombre_directorios = os.listdir(ruta_origen)

ruta_capas_copia = []
for directorio in nombre_directorios:
    ruta_bd = os.path.join(ruta_origen,directorio)
    arcpy.env.workspace = ruta_bd
    for dataset in arcpy.ListDatasets():
        for capa in arcpy.ListFeatureClasses(feature_dataset=dataset):
            ruta_origen = os.path.join(ruta_bd,dataset,capa)
            # * Esta se crea para realizar proceso de copia a dataset
            ruta_destino_final = os.path.join(ruta_destino,dataset)
            # * Para analisis se crea la siguiente variabla
            ruta_destino_copia = os.path.join(ruta_destino,dataset,capa)
            comparacion_origen = ruta_origen.split('\\')[9][4:]
            comparacion_destino = ruta_destino_copia.split('\\')[8][4:]
            if comparacion_origen == comparacion_destino:
                arcpy.management.CopyFeatures(ruta_origen, ruta_destino_final)
                print(""" Se Copia de:
                      {0}
                      a: 
                      {1}
                      """.format(ruta_origen,ruta_destino_final))
```
