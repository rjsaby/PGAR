<h4>Asignación de nombre de POMCA a cada registro de cada capa asociada al mismo</h4>

```python
ruta_directorio_bases_datos = r"D:\PUBLIC\PGAR\Resultados\4.PRO\PGAR\gdb_pomcas"

# ** Lista de todos los directorios de la ruta inicial
nombre_bases_datos = os.listdir(ruta_directorio_bases_datos)

# ** Creación de lista vacía para guardar posteriormente las rutas de cada una de las bases de datos
rutas_bd_pomcas = []

# ** Generación de listado de rutas
for directorio in nombre_bases_datos:
    if '.gdb' in directorio and 'POMCA_Unificada' not in directorio:
        ruta_bd = os.path.join(ruta_directorio_bases_datos,directorio)
        rutas_bd_pomcas.append(ruta_bd)
        
# ** ***** FEATURES CLASS *****
nombre_campo_pomca = 'nombre_pomca'
# ** Creación de campo para asiganción del nombre y registro por calculo del nombre del POMCA
# ** esto para reconocer, cuando se unifique, a qué POMCA pertenece cada registro
for espacios_trabajo in rutas_bd_pomcas:
    arcpy.env.workspace = espacios_trabajo
    nombres_feature_dataset = arcpy.ListDatasets()
    for datasets in nombres_feature_dataset:
        for capas in arcpy.ListFeatureClasses(feature_dataset=datasets):
            descripcion = arcpy.Describe(capas)
            if descripcion.featureType != "Annotation":
                # ? Se utiliza la función arcpy.ListFields con wilkcard, la cual filtra inmediatamente la existencia de lo mencionado en el wilkcard
                if arcpy.ListFields(capas, nombre_campo_pomca):
                    print("El campo existe, por lo tanto NO se actualiza la capa {0} de la BD {1}".format(capas, espacios_trabajo))
                else:
                    # ** Actualización de bases de datos
                    arcpy.management.AddField(capas, nombre_campo_pomca, 'TEXT', field_length = '100', field_alias = 'Nombre POMCA')
                    print("Se crea campo {0} en capa {1}".format(nombre_campo_pomca, capas))
                    # ** Se crea la varible que almacena el nombre de la base de datos
                    expresion_calcular = arcpy.env.workspace.split('\\')[7].split('.')[0]
                    arcpy.management.CalculateField(capas, nombre_campo_pomca, "'"+expresion_calcular+"'", 'PYTHON3')
                    print("Se actualiza la capa {0}, de la BD {1}".format(capas, espacios_trabajo))
                    
# ** ***** FEATURES TABLE *****
nombre_campo_pomca = 'nombre_pomca'
# ** Creación de campo para asiganción del nombre y registro por calculo del nombre del POMCA
# ** esto para reconocer, cuando se unifique, a qué POMCA pertenece cada registro
for espacios_trabajo in rutas_bd_pomcas:
    arcpy.env.workspace = espacios_trabajo
    for tabla in arcpy.ListTables():
        # ? Se utiliza la función arcpy.ListFields con wilkcard, la cual filtra inmediatamente la existencia de lo mencionado en el wilkcard
        if arcpy.ListFields(tabla, nombre_campo_pomca):
            print("El campo existe, por lo tanto NO se actualiza la tabla {0} de la BD {1}".format(tabla, espacios_trabajo))
        else:
            # ** Actualización de bases de datos
            arcpy.management.AddField(tabla, nombre_campo_pomca, 'TEXT', field_length = '100', field_alias = 'Nombre POMCA')
            print("Se crea campo {0} en tabla {1}".format(nombre_campo_pomca, tabla))
            # ** Se crea la varible que almacena el nombre de la base de datos
            expresion_calcular = arcpy.env.workspace.split('\\')[7].split('.')[0]
            arcpy.management.CalculateField(tabla, nombre_campo_pomca, "'"+expresion_calcular+"'", 'PYTHON3')
            print("Se actualiza la tabla {0}, de la BD {1}".format(tabla, espacios_trabajo))
```

```python
nombre_feature_dataset = {11:'GEOLOGIA',
                          12:'GEOMORFOLOGIA',
                          14:'SUELOS',
                          15:'RECURSO_HIDRICO_SUPERFICIAL',
                          16:'HIDROGEOLOGIA',
                          17:'GEOTECNIA',
                          19:'CLIMA',
                          20:'BIOTICO',
                          22:'POLITICO_ADMINISTRATIVO',
                          23:'ECONOMICO',
                          24:'SOCIOCULTURAL',
                          26:'GESTION_RIESGO',
                          27:'INDICES',
                          28:'SINTESIS_AMBIENTAL',
                          30:'ZONIFICACION_POMCA',
                          31:'AREAS_Y_ECOSISTEMAS_ESTRATEGICOS',
                          32:'AREAS_REGLAMENTACION_ESPECIAL',
                          33:'ANALISIS_FUNCIONAL',
                          34:'PROYECTO',
                          35:'INDICES_2027',
                          36:'PERFILES_GEOLOGICOS',
                          37:'PERFILES_GEOMORFOLOGICOS'}

estructura_campos = {# ? Geología
                    'ContactoGeologico':'GEOLOGIA',
                    'ContactoGeologico100k':'GEOLOGIA',
                    'ContactoFotoGeologico':'GEOLOGIA',
                    'ContactoGeologico_Fotogeo':'GEOLOGIA',
                    'ContactoGeologico_Geo100K':'GEOLOGIA',
                    'ContactoGeologico_25k':'GEOLOGIA',
                    'ContactoGeologico_Geo25K':'GEOLOGIA',
                    'CortesGeologicos':'GEOLOGIA',
                    'DatoEstructuralGeol':'GEOLOGIA',
                    'DatoEstructuralGeol_Fotogeo':'GEOLOGIA',
                    'DatoEstructuralGeol_25k':'GEOLOGIA',
                    'DatoEstructuralGeol_Geo25K':'GEOLOGIA',
                    'DatoEstructuralGeol_100k':'GEOLOGIA',
                    'DatoEstructuralGeol100k':'GEOLOGIA',
                    'DatoEstructuralGeol_Geo100K':'GEOLOGIA',
                    'DatoEstructuralGeol_Campo':'GEOLOGIA',
                    'Depositos':'GEOLOGIA',
                    'Disposicion_de_Rocas':'GEOLOGIA',
                    'DensidadFracturamiento':'GEOLOGIA',
                    'DiscontinuidadGeologica':'GEOLOGIA',
                    'DiscontinuidadGeologicaPT':'GEOLOGIA',
                    'discontinuidad_geologica_p':'GEOLOGIA',
                    'Estaciones':'GEOLOGIA',
                    'EstructuraLineal_Geologica_100k':'GEOLOGIA',
                    'EstructuraFallaLineam':'GEOLOGIA',
                    'EstructuraFallaLineamFoto':'GEOLOGIA',
                    'EstructuraFallaLineam_Fotogeo':'GEOLOGIA',
                    'EstructuraFallaLineam_25K':'GEOLOGIA',
                    'EstructuraFallaLineam_Geo25K':'GEOLOGIA',
                    'EstructuraFallaLineam100k':'GEOLOGIA',
                    'EstructuraFallaLineam_Geo100K':'GEOLOGIA',
                    'EstructuraFallaLineamRegional':'GEOLOGIA',
                    'EstructuraFallaLineamLocal':'GEOLOGIA',
                    'EstructuraPliegueLNBasica':'GEOLOGIA',
                    'EstructuraPliegueLN100k':'GEOLOGIA',
                    'EstructuraPliegueLN100k':'GEOLOGIA',
                    'EstructuraPliegueLN_25K':'GEOLOGIA',
                    'EstructuraPliegueLN_Geo25K':'GEOLOGIA',
                    'EstructuraPliegueLN_Geo100K':'GEOLOGIA',
                    'EstructuraPliegueLN':'GEOLOGIA',
                    'EstructuraPliegueLN_Fotogeo':'GEOLOGIA',
                    'EstructuraPliegueLNRegional':'GEOLOGIA',
                    'EstructuraPliegueLNLocal':'GEOLOGIA',
                    'EstructuraPlieguePT':'GEOLOGIA',
                    'EstructuraPlieguePT_100K':'GEOLOGIA',
                    'ExploracionGeotecnica':'GEOLOGIA',
                    'FotoGeologiaParaGeologiaBasica':'GEOLOGIA',
                    'FotoGeologiaBasica':'GEOLOGIA',
                    'Fotogeologia':'GEOLOGIA',
                    'FotoGeologiaParaUGS':'GEOLOGIA',
                    'FotoGeologiaSuperficial':'GEOLOGIA',
                    'GeologicoGeomorfologico':'GEOLOGIA',
                    'GeologiaRegional':'GEOLOGIA',
                    'MaterialesConstruccionPG':'GEOLOGIA',
                    'MaterialesConstruccionPT':'GEOLOGIA',
                    'MuestreoUnidadGeologicaSuperficial':'GEOLOGIA',
                    'Perfil':'GEOLOGIA',
                    'PerfilGeologico_Geo100K':'GEOLOGIA',
                    'PuntosControl':'GEOLOGIA',
                    'PuntosGeotecniaPT':'GEOLOGIA',
                    'RecursosNaturales':'GEOLOGIA',
                    'UnidadGeologica':'GEOLOGIA',
                    'UnidadGeologicaLocal':'GEOLOGIA',
                    'UnidadGeologica_25k':'GEOLOGIA',
                    'UnidadGeologica_Geo25K':'GEOLOGIA',
                    'UnidadGeologica100k':'GEOLOGIA',
                    'UnidadGeologica_100k':'GEOLOGIA',
                    'UnidadGeologica_Geo100K':'GEOLOGIA',                
                    'UnidadGeologicaRegional':'GEOLOGIA',
                    'UnidadGeologicaSuperficial':'GEOLOGIA',
                    'UnidadFotoGeologicaSuperficial':'GEOLOGIA',
                    'UnidadGeologicaSuperficial_Cuantitativa':'GEOLOGIA',
                    'UnidadFotoGeologica':'GEOLOGIA',
                    'UnidadGeologica_Fotogeo':'GEOLOGIA',
                    # ? Geomorfología
                    'FotointerpretacionGeomorfologica':'GEOMORFOLOGIA',
                    'FotoGeomorfologiaBasica':'GEOMORFOLOGIA',
                    'Pendiente':'GEOMORFOLOGIA',
                    'PendienteGR':'GEOMORFOLOGIA',
                    'PendienteGrados':'GEOMORFOLOGIA',
                    'PendientePorcentaje':'GEOMORFOLOGIA',
                    'Perfil_Geomorfologico':'GEOMORFOLOGIA',
                    'ProcesoMorfodinamicoLN':'GEOMORFOLOGIA',
                    'ProcesoMorfodinamicoPG':'GEOMORFOLOGIA',
                    'ProcesoMorfodinamicoPG_Geomorfo25K':'GEOMORFOLOGIA',
                    'ProcesoMorfodinamicoPT':'GEOMORFOLOGIA',
                    'ProcesoMorfodinamicoPT_Geomorfo25K':'GEOMORFOLOGIA',
                    'SubunidadGeomorfologica':'GEOMORFOLOGIA',
                    'UnidadGeomorfologicaIGAC':'GEOMORFOLOGIA',
                    'UnidadGeomorfologicaIGAC_RFP':'GEOMORFOLOGIA',
                    'UnidadGeomorfologicaIGAC_V1':'GEOMORFOLOGIA',
                    'UnidadGeomorfologicaIGAC_V3':'GEOMORFOLOGIA',
                    'UnidadGeomorfologicaSGC':'GEOMORFOLOGIA',
                    'UnidadGeomorfologicaSGC_Fotogeomorfo':'GEOMORFOLOGIA',
                    'UnidadGeomorfologicaSGC_Geomorfo25K':'GEOMORFOLOGIA',
                    'UnidadFotoGeomorfologicaSGC':'GEOMORFOLOGIA',
                    # ? Suelos
                    'CapacidadUsoTierra':'SUELOS',
                    'PuntoMuestreoEdafologia':'SUELOS',               
                    'PuntoMuestreoSuelo':'SUELOS',
                    'Suelo':'SUELOS',                    
                    # ? Recurso Hídrico Superficial
                    'CaptacionAguaSuperLN':'RECURSO_HIDRICO_SUPERFICIAL',
                    'CaptacionAguaSuperPT':'RECURSO_HIDRICO_SUPERFICIAL',
                    'Caudal':'RECURSO_HIDRICO_SUPERFICIAL',
                    'Caudales_Cuenca':'RECURSO_HIDRICO_SUPERFICIAL',
                    'CuencaHidrografica':'RECURSO_HIDRICO_SUPERFICIAL',
                    'CuencaHidrograficasubzonahidrografica':'RECURSO_HIDRICO_SUPERFICIAL',
                    'CaudalesMinimos':'RECURSO_HIDRICO_SUPERFICIAL',
                    'CaudalesMedios':'RECURSO_HIDRICO_SUPERFICIAL',
                    'CaudalesMaximos':'RECURSO_HIDRICO_SUPERFICIAL',
                    'Demanda':'RECURSO_HIDRICO_SUPERFICIAL',
                    'DemandaHidrica':'RECURSO_HIDRICO_SUPERFICIAL',
                    'DemandaHidricaTotal':'RECURSO_HIDRICO_SUPERFICIAL',
                    'DemandaHidricaDomestica':'RECURSO_HIDRICO_SUPERFICIAL',
                    'DemandaHidricaAgricola':'RECURSO_HIDRICO_SUPERFICIAL',
                    'EstacionesHidrometricas':'RECURSO_HIDRICO_SUPERFICIAL',
                    'MicroCuencas':'RECURSO_HIDRICO_SUPERFICIAL',
                    'microcuencasunidadhidrograficanivelI':'RECURSO_HIDRICO_SUPERFICIAL',
                    'Licencias_Corpoboyaca':'RECURSO_HIDRICO_SUPERFICIAL',
                    'MicrocuencaHidrografica':'RECURSO_HIDRICO_SUPERFICIAL',
                    'MicroCuencasAbastecedoras':'RECURSO_HIDRICO_SUPERFICIAL',
                    'OcupacionesCaucePG':'RECURSO_HIDRICO_SUPERFICIAL',
                    'PuntoMuestreoAguaSuper':'RECURSO_HIDRICO_SUPERFICIAL',
                    'RendimientoHidrico':'RECURSO_HIDRICO_SUPERFICIAL',
                    'RendimientoMinimo':'RECURSO_HIDRICO_SUPERFICIAL',
                    'RendimientoMedio':'RECURSO_HIDRICO_SUPERFICIAL',
                    'RendimientoMaximo':'RECURSO_HIDRICO_SUPERFICIAL',
                    'Rend_Hidrico_Minimo_Mensual_Anual':'RECURSO_HIDRICO_SUPERFICIAL',
                    'Rend_Hidrico_Medio_Mensual_Anual':'RECURSO_HIDRICO_SUPERFICIAL',
                    'Rend_Hidrico_Maximo_Mensual_Anual':'RECURSO_HIDRICO_SUPERFICIAL',                 
                    'SubCuencaHidrografica':'RECURSO_HIDRICO_SUPERFICIAL',
                    'Subcuencasunidadhidrograficanivel':'RECURSO_HIDRICO_SUPERFICIAL',
                    'Strahler':'RECURSO_HIDRICO_SUPERFICIAL',
                    'Unidades_Hidrográficas_Nivel_2_UHN_2':'RECURSO_HIDRICO_SUPERFICIAL',
                    'Unidades_Hidrográficas_Nivel_1_UHN_1':'RECURSO_HIDRICO_SUPERFICIAL',
                    'VertimientoLN':'RECURSO_HIDRICO_SUPERFICIAL',
                    'VertimientoPG':'RECURSO_HIDRICO_SUPERFICIAL',
                    'VertimientoPT':'RECURSO_HIDRICO_SUPERFICIAL',
                    # ? Hidrogeología
                    'LineasFlujo':'HIDROGEOLOGIA',
                    'Muestreo_UGS':'HIDROGEOLOGIA',
                    'PendienteParaHidrogeologia':'HIDROGEOLOGIA',
                    'PerdidaSello':'HIDROGEOLOGIA',
                    'Productividad':'HIDROGEOLOGIA',
                    'PuntoHidrogeologico':'HIDROGEOLOGIA',
                    'UnidadHidrogeologica':'HIDROGEOLOGIA',
                    'UnidadHidrogeologica1':'HIDROGEOLOGIA',
                    'UnidadHidrogeologica_25k':'HIDROGEOLOGIA',
                    'ZonasImportanciaHidrogeologica':'HIDROGEOLOGIA',
                    'ZonasProteccion':'HIDROGEOLOGIA',
                    'ZonasRecarga':'HIDROGEOLOGIA',
                    # ? Geotecnia
                    'ZonificacionGeotecnica':'GEOTECNIA',
                    # ? Clima
                    'EstacionMeteorologica':'CLIMA',
                    'Isoyeta':'CLIMA',
                    'Isoterma':'CLIMA',
                    'ZonificacionClimatica':'CLIMA',
                    'Evapotranspiración':'CLIMA',
                    'BalanceHidrico':'CLIMA',
                    'IsoyetaPG':'CLIMA',
                    'IsotermaPG':'CLIMA',
                    'Isoyeta_enero':'CLIMA',
                    'Isoyeta_febrero':'CLIMA',
                    'Isoyeta_marzo':'CLIMA',
                    'Isoyeta_abril':'CLIMA',
                    'Isoyeta_mayo':'CLIMA',
                    'Isoyeta_junio':'CLIMA',
                    'Isoyeta_julio':'CLIMA',
                    'Isoyeta_agosto':'CLIMA',
                    'Isoyeta_septiembre':'CLIMA',
                    'Isoyeta_octubre':'CLIMA',
                    'Isoyeta_noviembre':'CLIMA',
                    'Isoyeta_diciembre':'CLIMA',
                    'IsoyetaPG_enero':'CLIMA',
                    'IsoyetaPG_febrero':'CLIMA',
                    'IsoyetaPG_marzo':'CLIMA',
                    'IsoyetaPG_abril':'CLIMA',
                    'IsoyetaPG_mayo':'CLIMA',
                    'IsoyetaPG_junio':'CLIMA',
                    'IsoyetaPG_julio':'CLIMA',
                    'IsoyetaPG_agosto':'CLIMA',
                    'IsoyetaPG_septiembre':'CLIMA',
                    'IsoyetaPG_octubre':'CLIMA',
                    'IsoyetaPG_noviembre':'CLIMA',
                    'IsoyetaPG_diciembre':'CLIMA',
                    # ? Biotico
                    'CoberturaTierra_UsoActual':'BIOTICO',
                    'PuntoMuestreo_AvistamientoFauna':'BIOTICO',
                    'SitioImportanciaContinental':'BIOTICO',
                    'PuntoMuestreoFlora':'BIOTICO',
                    'MultitemporalCoberturas':'BIOTICO',
                    # ? Político - Administrativo
                    'Municipio':'POLITICO_ADMINISTRATIVO',
                    'Vereda':'POLITICO_ADMINISTRATIVO',
                    'Asentamiento':'POLITICO_ADMINISTRATIVO',
                    'Departamento':'POLITICO_ADMINISTRATIVO',
                    # ? Económico
                    'InfraestructuraLN':'ECONOMICO',
                    'InfraestructuraPT':'ECONOMICO',
                    'RutaMovilizacion':'ECONOMICO',
                    'EstructuraPropiedad':'ECONOMICO',
                    'FormaTenencia':'ECONOMICO',
                    'Predial':'ECONOMICO',
                    'MacroProyectosFuturos':'ECONOMICO',
                    # ? Sociocultural
                    'ProyeccionDesarrollo':'SOCIOCULTURAL',
                    'ReservaIndigena':'SOCIOCULTURAL',
                    'SabanaComunal':'SOCIOCULTURAL',
                    'ReservaCampesina':'SOCIOCULTURAL',
                    'ProyeccionDesarrolloPT':'SOCIOCULTURAL',
                    'DensidadPoblacional':'SOCIOCULTURAL',
                    'EscenDeseados_PG':'SOCIOCULTURAL',
                    'EscenDeseados_PT':'SOCIOCULTURAL',
                    # ? Gestión Riesgo
                    'AmenazaDesertizacion':'GESTION_RIESGO',
                    'AmenazaErosionCostera':'GESTION_RIESGO',
                    'AmenazaOtras':'GESTION_RIESGO',
                    'AmenazaSismica':'GESTION_RIESGO',
                    'AmenazaTsunami':'GESTION_RIESGO',
                    'AmenazaVolcanica':'GESTION_RIESGO',
                    'ElementosExpuestosPT':'GESTION_RIESGO',
                    'ElementosExpuestosLN':'GESTION_RIESGO',
                    'ElementosExpuestosPG':'GESTION_RIESGO',
                    'EscenAmenAvenTorren':'GESTION_RIESGO',
                    'EscenAmenIncendio':'GESTION_RIESGO',
                    'EscenAmenInunda':'GESTION_RIESGO',
                    'EscenAmenMovMasa':'GESTION_RIESGO',
                    'EscenRiesgoAvenTorren':'GESTION_RIESGO',
                    'EscenRiesgoIncendio':'GESTION_RIESGO',
                    'EscenRiesgoInundacion':'GESTION_RIESGO',
                    'EscenRiesgoMovMasa':'GESTION_RIESGO',
                    'EventosPT':'GESTION_RIESGO',
                    'EventosPG':'GESTION_RIESGO',
                    'Suscept_AvenTorren':'GESTION_RIESGO',
                    'Suscept_Incendios':'GESTION_RIESGO',
                    'Suscept_Inundaciones':'GESTION_RIESGO',
                    'Suscept_MovMasa':'GESTION_RIESGO',
                    'Vulnerabilidad_PT':'GESTION_RIESGO',
                    'Vulnerabilidad_LN':'GESTION_RIESGO',
                    'Vulnerabilidad_PG':'GESTION_RIESGO',
                    'ZonasHomogeneasCentrosPoblados':'GESTION_RIESGO',
                    'ZonasHomogeneasRurales':'GESTION_RIESGO',
                    'AmenazaInundaciones':'GESTION_RIESGO',
                    'AmenazaMovMasa':'GESTION_RIESGO',
                    'AmenazaAveTorren':'GESTION_RIESGO',
                    'VulnerabilidadAvenidasTorrenciales':'GESTION_RIESGO',
                    'VulnerabilidadInundacion':'GESTION_RIESGO',
                    'VulnerabilidadMovMasas':'GESTION_RIESGO',
                    'AmenazaIncendios':'GESTION_RIESGO',
                    'AmenazaIncendiosNiño':'GESTION_RIESGO',
                    'VulnerabilidadIncendForestales':'GESTION_RIESGO',
                    'ZonificacionMateriales':'GESTION_RIESGO',
                    'PoligonoDeslizamiento':'GESTION_RIESGO',
                    'PerfilesGeotecnicos':'GESTION_RIESGO',
                    # ? Índices
                    'Indicador_VegRemanente':'INDICES',
                    'IndicadorTasaCambio':'INDICES',
                    'IndicadorPresionDemog':'INDICES',
                    'IndiceAmbienteCritico':'INDICES',
                    'IndiceEstadoActCoberNat':'INDICES',
                    'IndiceCalidadAguaHumedo':'INDICES',
                    'IndiceCalidadAguaSeco':'INDICES',
                    'IndiceAltPotCalidadAgua':'INDICES',
                    'IndiceRetencionRegHidrica':'INDICES',
                    'IndiceAridez':'INDICES',
                    'IndiceUsoAguaSuperf':'INDICES',
                    'IndiceVulnerabDesabastHidrico':'INDICES',
                    'IndiceVulnerabEventTorren':'INDICES',
                    'IndiceExposición':'INDICES',
                    'IndiceFragilidad':'INDICES',
                    'IndiceResiliencia':'INDICES',
                    'ÍndiceFragmentacion':'INDICES',
                    'IndiceRetencionRegHidricaSub':'INDICES',
                    'IndiceUsoAguaSupSub':'INDICES',
                    'IndiceVulnerabDesabastHidricoSub':'INDICES',
                    # ? Síntesis Ambiental
                    'AreasCriticas':'SINTESIS_AMBIENTAL',
                    'ConflictoPerdidaCoberturasNaturales':'SINTESIS_AMBIENTAL',
                    'ConflictoUsoSuelo':'SINTESIS_AMBIENTAL',
                    'ConflictoUsoAgua':'SINTESIS_AMBIENTAL',
                    # ? Zonificación POMCA
                    'Zambiental_Paso1_POMCAS':'ZONIFICACION_POMCA',
                    'Zambiental_Paso2_POMCAS':'ZONIFICACION_POMCA',
                    'Zambiental_Paso3_POMCAS':'ZONIFICACION_POMCA',
                    'Zambiental_Paso4_POMCAS':'ZONIFICACION_POMCA',
                    'Zambiental_Paso5_POMCAS':'ZONIFICACION_POMCA',
                    'ZonificacionAmbiental_POMCAS':'ZONIFICACION_POMCA',
                    # ? Áreas y Ecosistemas
                    'AECC_AICAS':'AREAS_ECOSISTEMAS_ESTRATEGICOS',
                    'AECC_ReservaBiosfera':'AREAS_ECOSISTEMAS_ESTRATEGICOS',
                    'AECC_PatrimonioMundial':'AREAS_ECOSISTEMAS_ESTRATEGICOS',
                    'AECC_RAMSAR':'AREAS_ECOSISTEMAS_ESTRATEGICOS',
                    'AECC_ClaseSuelo':'AREAS_ECOSISTEMAS_ESTRATEGICOS',
                    'AECC_ReservaForestalLey2':'AREAS_ECOSISTEMAS_ESTRATEGICOS',
                    'AIA_EcoEstrat_Paramo':'AREAS_ECOSISTEMAS_ESTRATEGICOS',
                    'AIA_EcoEstrat_Humedal':'AREAS_ECOSISTEMAS_ESTRATEGICOS',
                    'AIA_OtrasAreas':'AREAS_ECOSISTEMAS_ESTRATEGICOS',
                    'SINAP_AreasRecreacion':'AREAS_ECOSISTEMAS_ESTRATEGICOS',
                    'SINAP_DistritoConservacionSuelos':'AREAS_ECOSISTEMAS_ESTRATEGICOS',
                    'SINAP_DistritoManejoIntegrado':'AREAS_ECOSISTEMAS_ESTRATEGICOS',
                    'SINAP_ParqueNatural':'AREAS_ECOSISTEMAS_ESTRATEGICOS',
                    'SINAP_ReservaForestalProtectora':'AREAS_ECOSISTEMAS_ESTRATEGICOS',
                    'SINAP_ReservaNaturalSociedadCivil':'AREAS_ECOSISTEMAS_ESTRATEGICOS',
                    'AP_Otras':'AREAS_ECOSISTEMAS_ESTRATEGICOS',
                    'AP_SDA':'AREAS_ECOSISTEMAS_ESTRATEGICOS',
                    # ? Área Reglamentación Especial
                    'ResguardoIndigena':'AREAS_REGLAMENTACION_ESPECIAL',
                    'SitioInteresCultural':'AREAS_REGLAMENTACION_ESPECIAL',
                    'TerritColectComNegras':'AREAS_REGLAMENTACION_ESPECIAL',
                    'ZonaArqueologica':'AREAS_REGLAMENTACION_ESPECIAL',
                    # ? Análisis Funcional
                    'AreasDeServicio':'ANALISIS_FUNCIONAL',
                    'Cuencas Adyacente':'ANALISIS_FUNCIONAL',
                    'DensidadPorMunicipios':'ANALISIS_FUNCIONAL',
                    'MunicipiosAdyacente':'ANALISIS_FUNCIONAL',
                    'RedVial':'ANALISIS_FUNCIONAL',
                    'SubsistemaFuncionalLT':'ANALISIS_FUNCIONAL',
                    'SubsistemaFuncionalesPL':'ANALISIS_FUNCIONAL',
                    'TipoDeCentralidad':'ANALISIS_FUNCIONAL',
                    'AreasPreservacionAntena':'ANALISIS_FUNCIONAL',
                    'AreasPreservacionCentraldeEnergia':'ANALISIS_FUNCIONAL',
                    'AreasPreservacionConstruccionPT':'ANALISIS_FUNCIONAL',
                    'AreasPreservacionPlantaTratamiento':'ANALISIS_FUNCIONAL',
                    'AreasPreservacionRedAltaTension':'ANALISIS_FUNCIONAL',
                    'AreasPreservacionTorre':'ANALISIS_FUNCIONAL',
                    # ? Proyecto
                    'LimiteDepartamentos':'PROYECTO',
                    'LimiteColombia':'PROYECTO',
                    'LimiteCuenca':'PROYECTO',
                    'Indice25K':'PROYECTO',
                    'Indice100K':'PROYECTO',
                    # ? Índices 2027
                    'Indicador_VegRemanente_2027':'INDICES_2027',
                    'IndicadorTasaCambio_2027':'INDICES_2027',
                    'IndicadorPresionDemog_2027':'INDICES_2027',
                    'IndiceAmbienteCritico_2027':'INDICES_2027',
                    'IndiceEstadoActCoberNat_2027':'INDICES_2027',
                    'IndiceFragmentacion_2027':'INDICES_2027',
                    'IndiceAridez_2027':'INDICES_2027',
                    # ? Perfiles Geológicos
                    'FALLAS_AA':'PERFILES_GEOLOGICOS',
                    'FALLAS_BB':'PERFILES_GEOLOGICOS',
                    'FALLAS_CC':'PERFILES_GEOLOGICOS',
                    'FALLAS_DD':'PERFILES_GEOLOGICOS',
                    'FALLAS_EE':'PERFILES_GEOLOGICOS',
                    'FALLAS_FF':'PERFILES_GEOLOGICOS',
                    'FALLAS_GG':'PERFILES_GEOLOGICOS',
                    'GEOLOGIA_BASICA_AA_LN':'PERFILES_GEOLOGICOS',
                    'GEOLOGIA_BASICA_BB_LN':'PERFILES_GEOLOGICOS',
                    'GEOLOGIA_BASICA_CC_LN':'PERFILES_GEOLOGICOS',
                    'GEOLOGIA_BASICA_DD_LN':'PERFILES_GEOLOGICOS',
                    'GEOLOGIA_BASICA_EE_LN':'PERFILES_GEOLOGICOS',
                    'GEOLOGIA_BASICA_FF_LN':'PERFILES_GEOLOGICOS',
                    'GEOLOGIA_BASICA_GG_LN':'PERFILES_GEOLOGICOS',
                    'GEOLOGIA_UGS_AA_LN':'PERFILES_GEOLOGICOS',
                    'GEOLOGIA_UGS_BB_LN':'PERFILES_GEOLOGICOS',
                    'GEOLOGIA_UGS_CC_LN':'PERFILES_GEOLOGICOS',
                    'GEOLOGIA_UGS_DD_LN':'PERFILES_GEOLOGICOS',
                    'GEOLOGIA_UGS_EE_LN':'PERFILES_GEOLOGICOS',
                    'GEOLOGIA_UGS_FF_LN':'PERFILES_GEOLOGICOS',
                    'GEOLOGIA_UGS_GG_LN':'PERFILES_GEOLOGICOS',                    
                    'GEOLOGIA_BASICA_AA':'PERFILES_GEOLOGICOS',
                    'GEOLOGIA_BASICA_BB':'PERFILES_GEOLOGICOS',
                    'GEOLOGIA_BASICA_CC':'PERFILES_GEOLOGICOS',
                    'GEOLOGIA_BASICA_DD':'PERFILES_GEOLOGICOS',
                    'GEOLOGIA_BASICA_EE':'PERFILES_GEOLOGICOS',
                    'GEOLOGIA_BASICA_FF':'PERFILES_GEOLOGICOS',
                    'GEOLOGIA_BASICA_GG':'PERFILES_GEOLOGICOS',
                    'GEOLOGIA_UGS_AA':'PERFILES_GEOLOGICOS',
                    'GEOLOGIA_UGS_BB':'PERFILES_GEOLOGICOS',
                    'GEOLOGIA_UGS_CC':'PERFILES_GEOLOGICOS',
                    'GEOLOGIA_UGS_DD':'PERFILES_GEOLOGICOS',
                    'GEOLOGIA_UGS_EE':'PERFILES_GEOLOGICOS',
                    'GEOLOGIA_UGS_FF':'PERFILES_GEOLOGICOS',
                    'GEOLOGIA_UGS_GG':'PERFILES_GEOLOGICOS',
                    'GEOLOGIA_INDICE_CC':'PERFILES_GEOLOGICOS',
                    # ? Perfiles Geomorfológicos
                    'GEOMOR_IGAC_AA_LN':'PERFILES_GEOMORFOLOGICOS',
                    'GEOMOR_IGAC_BB_LN':'PERFILES_GEOMORFOLOGICOS',
                    'GEOMOR_IGAC_CC_LN':'PERFILES_GEOMORFOLOGICOS',
                    'GEOMOR_IGAC_DD_LN':'PERFILES_GEOMORFOLOGICOS',
                    'GEOMOR_IGAC_EE_LN':'PERFILES_GEOMORFOLOGICOS',
                    'GEOMOR_IGAC_FF_LN':'PERFILES_GEOMORFOLOGICOS',
                    'GEOMOR_IGAC_GG_LN':'PERFILES_GEOMORFOLOGICOS',
                    'GEOMOR_IGAC_AA':'PERFILES_GEOMORFOLOGICOS',
                    'GEOMOR_IGAC_BB':'PERFILES_GEOMORFOLOGICOS',
                    'GEOMOR_IGAC_CC':'PERFILES_GEOMORFOLOGICOS',
                    'GEOMOR_IGAC_DD':'PERFILES_GEOMORFOLOGICOS',
                    'GEOMOR_IGAC_EE':'PERFILES_GEOMORFOLOGICOS',
                    'GEOMOR_IGAC_FF':'PERFILES_GEOMORFOLOGICOS',
                    'GEOMOR_IGAC_GG':'PERFILES_GEOMORFOLOGICOS',
                    'GEOMOR_INDICE_CC':'PERFILES_GEOMORFOLOGICOS',
                    'GEOMOR_SGC_AA_LN':'PERFILES_GEOMORFOLOGICOS',
                    'GEOMOR_SGC_BB_LN':'PERFILES_GEOMORFOLOGICOS',
                    'GEOMOR_SGC_CC_LN':'PERFILES_GEOMORFOLOGICOS',
                    'GEOMOR_SGC_DD_LN':'PERFILES_GEOMORFOLOGICOS',
                    'GEOMOR_SGC_EE_LN':'PERFILES_GEOMORFOLOGICOS',
                    'GEOMOR_SGC_FF_LN':'PERFILES_GEOMORFOLOGICOS',
                    'GEOMOR_SGC_GG_LN':'PERFILES_GEOMORFOLOGICOS',
                    'GEOMOR_SGC_AA':'PERFILES_GEOMORFOLOGICOS',
                    'GEOMOR_SGC_BB':'PERFILES_GEOMORFOLOGICOS',
                    'GEOMOR_SGC_CC':'PERFILES_GEOMORFOLOGICOS',
                    'GEOMOR_SGC_DD':'PERFILES_GEOMORFOLOGICOS',
                    'GEOMOR_SGC_EE':'PERFILES_GEOMORFOLOGICOS',
                    'GEOMOR_SGC_FF':'PERFILES_GEOMORFOLOGICOS',
                    'GEOMOR_SGC_GG':'PERFILES_GEOMORFOLOGICOS'
                    }

estructura_tablas = {1:'MuestreoSueloQuimicasTB',
                     2:'MuestreoSueloFisicasTB',
                     3:'MuestreoFisicoquimSuperTB',
                     4:'MuestreoHidrobioTB',
                     5:'CaracterizaSedimentoTB',
                     6:'MuestreoFisicoquimSubterTB',
                     7:'SondeoElectricoVerticalTB',
                     8:'ContFuenteFijaEmisionTB',
                     9:'ContFuenteDispEmisionTB',
                     10:'FuentesMovilesEmisionTB',
                     11:'RegistrosCalidadAireTB',
                     12:'RegistrosRuidoAmbientalTB',
                     13:'RegistrosEstMeteorologicaTB',
                     14:'MuestreoFloraTB',
                     15:'MuestreoFaunaTB',
                     16:'AprovechaForestalTB',
                     17:'MuestreoFloraMarinaTB',
                     18:'MuestreoFaunaMarinaTB',
                     19:'MuestreoFisicoquimMarinoTB',
                     20:'MuestreoHidrobioMarinoTB',
                     21:'TendenciaEsperadaTB',
                     22:'RegistrosMultimediaTB',
                     23:'MuestreoFaunaOtrosLugaresTB',
                     24:'MuestreoFaunaEntrevistasTB',
                     25:'CoberturasTB',
                     26:'BocatomasCentrosPobladosTB'
                     }

def ejecucion_unificacion(wilkcard_dataframe):
        
    arcpy.env.workspace = ruta_bd_unificada

    # * Se almacena en una lista los dataframes
    # * esto se hace para posteriormente saber si se crean los dataframe o se truncan las capas alojadas
    print("1. Inicia generación de listado de feature datasets")
    lista_datasets = []
    for dataframe in arcpy.arcpy.ListDatasets():
        lista_datasets.append(dataframe)
    print("2. Finaliza generación de listado de feature datasets")
        
    # * Teniendo en cuenta la lista previamente creada, se crean o truncan los
    # * datos con las funciones a continuación expuestas
    print("3. Inicia borrado de feature class")
    if len(lista_datasets) != 0:
        for dataframe in arcpy.arcpy.ListDatasets(wild_card=wilkcard_dataframe):
            for capas in arcpy.ListFeatureClasses(feature_dataset=dataframe):
                arcpy.management.TruncateTable(capas)
                print(" - Se trunca la capa: {0}".format(capas))
        # !! tratamiento de truncado de tablas, aquí por favor Rodian
    else:
        print("Sin dataset")
    print("4. Finaliza borrado de feature class")
     
    # * Se trae la lista donde se almacenaron las rutas de cada una de las bases de datos de POMCAS
    # TODO: Bluque para el manejo de los feature classes
    print("5. Inicia generación de listados vacíos")
    for rutas in rutas_bd_pomcas:
        arcpy.env.workspace = rutas
        for ds in arcpy.ListDatasets(wild_card=wilkcard_dataframe):
            for capas in arcpy.ListFeatureClasses(feature_dataset=ds):
                # ? Por medio de la función globals, y complemento, se crean variables globales.
                # !! Revisar funcionamiento completo de la función en wiki
                globals()['lista%s' % capas] = []
    print("6. Finaliza generación de listados vacíos")
     
    # * Este blucle, mediante la función append, llena cada una de las listas creadas con variables globales
    # * con las rutas asociadas a cada feature dataset
    # TODO: Esto se realiza para los feature class
    print("7. Inicia construcción de rutas en listas creadas")
    for rutas in rutas_bd_pomcas:
        arcpy.env.workspace = rutas
        for ds in arcpy.ListDatasets(wild_card=wilkcard_dataframe):
            for capas in arcpy.ListFeatureClasses(feature_dataset=ds):
                for llave, valor in estructura_campos.items():
                    if llave == capas:
                        ruta = os.path.join(arcpy.env.workspace,ds,capas)
                        globals()['lista%s' % capas].append(ruta)
    print("8. Finaliza construcción de rutas en listas creadas")
                               
    # # * Se cambio el espacio de trabajo                 
    arcpy.env.workspace = ruta_bd_unificada

    print("9. Inicia adición de registros")
    for ds in arcpy.ListDatasets(wild_card=wilkcard_dataframe):
        for capas in arcpy.ListFeatureClasses(feature_dataset=ds):
            for llave, valor in estructura_campos.items():
                if llave == capas:
                    arcpy.management.Append(globals()['lista%s' % capas], capas, 'NO_TEST')
                    print(' - Se incorporan datos en la capa {0}'.format(capas))
                # else:
                #     print('Las capa {0} no coincide con la estructura'.format(capas))
    print("10. Finaliza adición de registros")

def unificacion_tablas():
    
    # * Bucle para borrado de tablas    
    for tabla in arcpy.ListTables():
        arcpy.management.TruncateTable(tabla)
        
    # TODO: Bluque para el manejo de los feature tables. Aquí se crearan las variables globales para las tablas    
    for rutas in rutas_bd_pomcas:
        arcpy.env.workspace = rutas
        for tabla in arcpy.ListTables():
            # ? Por medio de la función globals, y complemento, se crean variables globales.
            globals()['lista%s' % tabla] = []
            
    # TODO: Esto se realiza para los feature tables
    for rutas in rutas_bd_pomcas:
        arcpy.env.workspace = rutas
        for tabla in arcpy.ListTables():
            for llave, valor in estructura_campos.items():
                    if valor == tabla:
                        ruta_tabla = os.path.join(arcpy.env.workspace, tabla)
                        globals()['lista%s' % tabla].append(ruta_tabla)
                        
   # * Se cambio el espacio de trabajo                 
    arcpy.env.workspace = ruta_bd_unificada
    
    # * Se carga la información de cada ruta, de cada gdb, a la tabla correspondiente dentro de la bd unificada.
    for tabla in arcpy.ListTables():
        for llave, valor in estructura_campos.items():
            if valor == tabla:
                arcpy.management.Append(globals()['lista%s' % tabla], tabla, 'NO_TEST')
                print('Se incorporan datos en la capa {0}'.format(tabla))
            else:
                print('Las capa {0} no coincide con la estructura'.format(tabla))
```
